<p>The figures presented here do not take into account differences in the cost of living in different countries, and the results vary greatly from one year to another based on fluctuations in the exchange rates of the country's currency. Such fluctuations change a country's ranking from one year to the next, even though they often make little or no difference to the standard of living of its population.

GDP per capita is often considered an indicator of a country's standard of living; however, this is inaccurate because GDP per capita is not a measure of personal income.

Comparisons of national income are also frequently made on the basis of purchasing power parity (PPP), to adjust for differences in the cost of living in different countries. (See List of countries by GDP (PPP) per capita.) PPP largely removes the exchange rate problem but not others; it does not reflect the value of economic output in international trade, and it also requires more estimation than GDP per capita. On the whole, PPP per capita figures are more narrowly spread than nominal GDP per capita figures.</p>

## groupby()

[Learn more about groupby](https://www.geeksforgeeks.org/pandas-groupby/)

```python
# ============================================
# GROUPBY EXPLORATION – GDP PER CAPITA DATA
# ============================================

# Select the GDP per capita column (numeric)
gdp_col = df.select_dtypes("number").columns[0]
```

```python
# --------------------------------------------
# Title: Average GDP per Capita by UN Region
# Explanation:
# Groups countries by UN region and calculates
# the average GDP per capita for each region.
# This helps compare overall regional wealth.
# --------------------------------------------

df.groupby("UN_Region")[gdp_col].mean().sort_values(ascending=False)
```

```python
# --------------------------------------------
# Title: Median GDP per Capita by UN Region
# Explanation:
# Calculates the median GDP per capita per region,
# which reduces the impact of extreme outliers
# such as very high-income countries.
# --------------------------------------------

df.groupby("UN_Region")[gdp_col].median().sort_values(ascending=False)
```

```python
# --------------------------------------------
# Title: Number of Countries per UN Region
# Explanation:
# Counts how many countries are included in
# each UN region to understand dataset coverage.
# --------------------------------------------

df.groupby("UN_Region")["Country/Territory"].count().sort_values(ascending=False)
```

```python
# --------------------------------------------
# Title: GDP Statistics per UN Region
# Explanation:
# Uses aggregation to calculate minimum,
# maximum, mean, and median GDP per capita
# for each UN region in one summary table.
# --------------------------------------------

df.groupby("UN_Region")[gdp_col].agg(["min", "max", "mean", "median"])
```

```python
# --------------------------------------------
# Title: Richest Country in Each UN Region
# Explanation:
# Identifies the country with the highest
# GDP per capita within each UN region.
# --------------------------------------------

df.loc[
    df.groupby("UN_Region")[gdp_col].idxmax(),
    ["UN_Region", "Country/Territory", gdp_col]
].sort_values(gdp_col, ascending=False)
```

```python
# --------------------------------------------
# Title: Poorest Country in Each UN Region
# Explanation:
# Identifies the country with the lowest
# GDP per capita within each UN region.
# --------------------------------------------

df.loc[
    df.groupby("UN_Region")[gdp_col].idxmin(),
    ["UN_Region", "Country/Territory", gdp_col]
].sort_values(gdp_col)
```

```python
# --------------------------------------------
# Title: GDP Rank Within Each UN Region
# Explanation:
# Ranks countries by GDP per capita within
# their own region, allowing comparison
# between countries in the same region.
# --------------------------------------------

df["Region_GDP_Rank"] = (
    df.groupby("UN_Region")[gdp_col]
      .rank(ascending=False, method="dense")
)

df.sort_values(["UN_Region", "Region_GDP_Rank"]).head(15)
```

## Filling 0 Values by average

```python
import numpy as np
```

```python
# replace 0 with null values
estimate_cols = ["WorldBank_Estimate", "UN_Estimate", "IMF_Estimate"]

df[estimate_cols] = df[estimate_cols].replace(0, np.nan)
```

```python
# Calculate the average of 'WorldBank_Estimate' and 'UN_Estimate' columns
df["avg_worldbank_un"] = df[["WorldBank_Estimate", "UN_Estimate"]].mean(axis=1)
```

```python
# Fill the null values in 'IMF_Estimate' column with the calculated average
df["IMF_Estimate"] = df["IMF_Estimate"].fillna(df["avg_worldbank_un"])
```

```python
# Drop the temporary 'avg_worldbank_un' column if not needed
df.drop(columns=["avg_worldbank_un"], inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --------------------------------------------
# Title: Bar Chart - Number of Countries per UN Region
# Explanation:
# Counts how many countries appear in each UN region.
# Bar charts are useful for comparing totals across categories.
# --------------------------------------------

region_counts = df["UN_Region"].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(region_counts.index, region_counts.values)
plt.title("Number of Countries per UN Region")
plt.xlabel("UN Region")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Histogram - Distribution of IMF GDP per Capita
# Explanation:
# Shows how IMF GDP per capita values are distributed.
# Histograms help reveal typical ranges, skew, and spread.
# --------------------------------------------

imf_col = [c for c in df.select_dtypes("number").columns if "IMF" in c][0]

plt.figure(figsize=(10, 5))
plt.hist(df[imf_col].dropna(), bins=30)
plt.title("Distribution of IMF GDP per Capita")
plt.xlabel("GDP per Capita (IMF Estimate)")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Box Plot - IMF GDP per Capita by UN Region
# Explanation:
# Compares the spread of IMF GDP per capita across regions.
# Box plots show medians, quartiles, and potential outliers.
# --------------------------------------------

regions = df["UN_Region"].dropna().unique()
data_by_region = [
    df.loc[df["UN_Region"] == r, imf_col].dropna().values
    for r in regions
]

plt.figure(figsize=(12, 6))
plt.boxplot(data_by_region, labels=regions, showfliers=True)
plt.title("IMF GDP per Capita by UN Region (Box Plot)")
plt.xlabel("UN Region")
plt.ylabel("GDP per Capita (IMF Estimate)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Scatter Plot - World Bank vs IMF GDP per Capita
# Explanation:
# Compares GDP per capita estimates between World Bank and IMF.
# Scatter plots help identify relationships and agreement between measures.
# --------------------------------------------

worldbank_col = [c for c in df.select_dtypes("number").columns if "World" in c or "Bank" in c][0]

scatter_df = df[[worldbank_col, imf_col]].dropna()

plt.figure(figsize=(7, 7))
plt.scatter(scatter_df[worldbank_col], scatter_df[imf_col])
plt.title("World Bank vs IMF GDP per Capita")
plt.xlabel("World Bank Estimate")
plt.ylabel("IMF Estimate")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Pie Chart - Proportion of Countries by UN Region
# Explanation:
# Shows the proportion of countries in each UN region.
# Pie charts are useful for visualising relative shares.
# --------------------------------------------

region_counts = df["UN_Region"].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(region_counts.values, labels=region_counts.index, autopct="%1.1f%%")
plt.title("Proportion of Countries by UN Region")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Line Chart - Top 20 Countries by IMF GDP per Capita
# Explanation:
# Displays the top 20 countries ranked by IMF GDP per capita.
# Line charts can show ordered trends across ranked data.
# --------------------------------------------

imf_col = [c for c in df.select_dtypes("number").columns if "IMF" in c][0]

top20_imf = df.sort_values(imf_col, ascending=False).head(20)

plt.figure(figsize=(10, 5))
plt.plot(top20_imf["Country/Territory"], top20_imf[imf_col], marker="o")
plt.title("Top 20 Countries by IMF GDP per Capita")
plt.xlabel("Country")
plt.ylabel("GDP per Capita (IMF Estimate)")
plt.xticks(rotation=75, ha="right")
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Stacked Bar Chart - IMF GDP Above vs Below World Average by Region
# Explanation:
# Compares how many countries per region are above
# or below the IMF world average GDP per capita.
# --------------------------------------------
imf_world_avg = df[imf_col].mean()
df["Above_World_Avg"] = df[imf_col] >= imf_world_avg
stacked_counts = (
    df.groupby(["UN_Region", "Above_World_Avg"])
      .size()
      .unstack(fill_value=0)
)
plt.figure(figsize=(10, 6))
plt.bar(stacked_counts.index, stacked_counts[True], label="Above Average")
plt.bar(
    stacked_counts.index,
    stacked_counts[False],
    bottom=stacked_counts[True],
    label="Below Average"
)
plt.title("IMF GDP per Capita Above vs Below World Average by Region")
plt.xlabel("UN Region")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Title: Violin Plot - IMF GDP per Capita Distribution by Region
# Explanation:
# Shows the distribution shape of IMF GDP per capita
# for each region, combining spread and density.
# --------------------------------------------

regions = df["UN_Region"].unique()
data = [
    df.loc[df["UN_Region"] == r, imf_col].dropna().values
    for r in regions
]

plt.figure(figsize=(12, 6))
plt.violinplot(data, showmeans=True, showmedians=True)
plt.xticks(range(1, len(regions) + 1), regions, rotation=45, ha="right")
plt.title("IMF GDP per Capita Distribution by UN Region")
plt.xlabel("UN Region")
plt.ylabel("GDP per Capita (IMF Estimate)")
plt.tight_layout()
plt.show()
```
